```python
fix_tiger2 = cv2.resize(fix_tiger, (404, 540)) # height and width cut to half
plt.imshow(fix_tiger2)
```

```python
flip_tiger = cv2.flip(fix_tiger, 0) # flipping the image
plt.imshow(flip_tiger)
```

```python
mirror_tiger = cv2.flip(fix_tiger, 1) # mirroring the image
plt.imshow(mirror_tiger)
```

```python
cv2.imwrite('mirror_tiger.jpg', mirror_tiger) # saving the image in BGR format
```

```python
mirror_tiger_rgb = cv2.cvtColor(mirror_tiger, cv2.COLOR_BGR2RGB) # saving the image in RGB format

cv2.imwrite("mirror_tiger.jpg", mirror_tiger_rgb) 
```

# VIDEOS

```python
# the given code can be used for capturing videos through the laptop camera
import cv2
cap = cv2.VideoCapture(0)
width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

while True:
    ret, frame = cap.read()
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
